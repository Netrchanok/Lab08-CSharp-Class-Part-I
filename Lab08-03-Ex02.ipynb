## คำถาม

1. Field ในภาษา C# คืออะไร


*คำตอบ*

Field ในภาษา C# คือ ตัวแปร (Variable) ที่ถูกประกาศไว้ภายในคลาส (Class) หรือโครงสร้าง (Struct) เพื่อเก็บค่าหรือข้อมูลของอ็อบเจ็กต์ (Object)

🔹 Field จะเป็นส่วนหนึ่งของ Object และสามารถเก็บค่าได้ตลอดอายุของ Object
🔹 Field สามารถมี Modifiers เช่น public, private, static, readonly ได้

2. Field มีหน้าที่อะไรใน class 


*คำตอบ*

 หน้าที่ของ Field ใน Class
1️⃣ เก็บสถานะ (State) ของ Object

Field ถูกใช้เพื่อเก็บค่าข้อมูลที่เกี่ยวข้องกับอ็อบเจ็กต์ เช่น ชื่อ, อายุ, เงินเดือน

ข้อมูลใน Field เป็น คุณสมบัติของอ็อบเจ็กต์ และสามารถเปลี่ยนแปลงได้

2️⃣ ช่วยให้ Object มีข้อมูลเฉพาะตัว

แต่ละอ็อบเจ็กต์ที่ถูกสร้างจาก Class สามารถมีค่าของ Field ที่แตกต่างกันได้

ทำให้สามารถจำลองสิ่งต่าง ๆ ในโลกความเป็นจริงได้ดีขึ้น

3️⃣ ใช้ในการคำนวณหรือประมวลผลข้อมูลใน Class

Field สามารถเก็บค่าที่ใช้ในการคำนวณหรือใช้ในเมธอดของคลาสได้

4️⃣ ใช้แชร์ข้อมูลระหว่างเมธอดภายในคลาส

Field สามารถถูกอ่านหรือเขียนจากเมธอดต่าง ๆ ภายในคลาสเดียวกัน

5️⃣ ใช้เป็นค่าเริ่มต้นของอ็อบเจ็กต์

สามารถกำหนดค่าเริ่มต้นให้ Field ได้ เมื่ออ็อบเจ็กต์ถูกสร้าง

3. การประกาศ Field ทำได้อย่างไร ให้อธิบายรูปแบบการประกาศ field ในคลาส


*คำตอบ*

รูปแบบการประกาศ Field ในคลาส (Class) ในภาษา C#
Field คือ ตัวแปรที่ถูกประกาศภายใน Class หรือ Struct เพื่อใช้เก็บค่าข้อมูลของอ็อบเจ็กต์ (Object)

🔹 รูปแบบการประกาศ Field ใน C#
csharp
Copy
Edit
<Access Modifier> <Field Type> <Field Name> [= Initial Value];
🔹 ส่วนประกอบของ Field

<Access Modifier> → กำหนดการเข้าถึง เช่น public, private, protected, internal

<Field Type> → ประเภทของข้อมูล เช่น int, string, double, bool

<Field Name> → ชื่อของ Field

= Initial Value (ไม่จำเป็นต้องมี) → กำหนดค่าเริ่มต้นให้ Field

4. Field และ Variable แตกต่างกันอย่างไร


5. เมื่อไหร่ควรใช้ Field และเมื่อไหร่ควรใช้ Variable

*คำตอบ*

เมื่อไหร่ควรใช้ Field และเมื่อไหร่ควรใช้ Variable?
🔹 Field และ Variable มีจุดประสงค์การใช้งานที่ต่างกัน
👉 Field ใช้เก็บข้อมูลของ Object หรือ Class
👉 Variable ใช้เก็บค่าชั่วคราวภายใน Method หรือ Block

📌 ใช้ Field เมื่อไหร่?
✅ 1. เมื่อข้อมูลเป็นส่วนหนึ่งของ Object หรือ Class

ข้อมูลที่เกี่ยวข้องกับสถานะ (State) ของอ็อบเจ็กต์

เช่น ชื่อ, อายุ, คะแนน, เงินเดือน

✅ 2. เมื่อข้อมูลต้องใช้ได้หลายที่ในคลาสเดียวกัน

Field สามารถใช้งานได้ในทุกเมธอดของคลาส

Variable ใช้ได้แค่ใน Method ที่ประกาศ

✅ 3. เมื่อข้อมูลต้องอยู่กับ Object ตลอดอายุของมัน

เช่น Player.hp, Car.speed ที่ต้องคงอยู่แม้เมธอดจบการทำงาน

✅ 4. เมื่อข้อมูลต้องการ Access Modifier (private, public, protected)

Field สามารถควบคุมการเข้าถึงได้ด้วย private, public, protected

Variable ไม่สามารถใช้ Access Modifier ได้

✅ 5. เมื่อใช้ static, readonly, หรือ const

static 👉 ใช้ Field ที่ต้องแชร์ค่าร่วมกันในทุกอ็อบเจ็กต์

readonly 👉 ใช้ Field ที่กำหนดค่าได้ครั้งเดียวใน Constructor

const 👉 ใช้ Field ที่เปลี่ยนแปลงค่าไม่ได้เลย

ใช้ Variable เมื่อไหร่?
✅ 1. เมื่อข้อมูลใช้แค่ชั่วคราวภายใน Method

ไม่ต้องการให้ค่าคงอยู่หลังจากเมธอดจบการทำงาน

✅ 2. เมื่อข้อมูลใช้แค่ภายใน Block หรือ Loop

เช่น ตัวนับ Loop for (int i = 0; i < 10; i++)

✅ 3. เมื่อข้อมูลไม่ต้องแชร์ระหว่าง Method หรือ Object

ตัวแปรใช้เฉพาะใน Method ที่เรียกใช้ ไม่ต้องเก็บไว้

✅ 4. เมื่อข้อมูลต้องการคำนวณหรือเปลี่ยนแปลงบ่อยๆ

เช่น ผลรวม, ค่าที่เปลี่ยนแปลงตามการทำงาน
