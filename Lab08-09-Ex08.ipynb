## คำถาม

1. static class คืออะไร และแตกต่างจาก class ธรรมดาอย่างไร

*คำตอบ*

3. สมาชิกทั้งหมดของ static class ต้องเป็น static หรือไม่ เพราะอะไร

*คำตอบ*

สรุป:
สมาชิกของ static class ต้องเป็น static ทุกตัว เนื่องจาก static class ไม่มีอินสแตนซ์ (ไม่สามารถสร้างอ็อบเจ็กต์) และสมาชิกของคลาสต้องสามารถเข้าถึงได้โดยตรงจากชื่อของคลาสโดยไม่ต้องใช้อินสแตนซ์

การใช้สมาชิก static ทำให้การทำงานของ static class สามารถรักษาความสมบูรณ์ในการออกแบบที่ไม่พึ่งพาสถานะของอ็อบเจ็กต์.

4. เราสามารถสร้าง instance ของ static class ได้หรือไม่  เพราะอะไร

*คำตอบ*

ไม่สามารถสร้างอินสแตนซ์ (instance) ของ static class ได้ ในภาษา C# เนื่องจากเหตุผลดังนี้:

เหตุผล:
static class ถูกออกแบบมาเพื่อไม่ให้มีอินสแตนซ์:

เมื่อประกาศคลาสเป็น static นั่นหมายความว่า คลาสนั้นไม่สามารถมีสถานะ (state) ที่เกี่ยวข้องกับอ็อบเจ็กต์ และ ไม่จำเป็นต้องมีอินสแตนซ์ เพื่อใช้ฟังก์ชันหรือข้อมูลในคลาสนั้น

การสร้างอินสแตนซ์ ของคลาสจะไม่สามารถทำได้ เพราะการสร้างอินสแตนซ์จะต้องมีการใช้เมมเบอร์ (สมาชิก) ที่ไม่เป็น static ซึ่ง static class จะไม่สามารถมีเมมเบอร์ที่ไม่เป็น static ได้

ไม่มีคอนสตรัคเตอร์ (Constructor) สำหรับ static class:

static class จะไม่มีคอนสตรัคเตอร์ (constructor) เพื่อใช้ในการสร้างอินสแตนซ์ของมัน

การพยายามสร้างอินสแตนซ์จาก static class เช่น new StaticClass() จะทำให้เกิดข้อผิดพลาด เพราะคลาสนั้นไม่สามารถถูกสร้างขึ้นจากคอนสตรัคเตอร์ได้
