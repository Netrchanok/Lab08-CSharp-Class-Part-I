## คำถาม



1. Class ในภาษา C# คืออะไร


*คำตอบ*

Class ในภาษา C# คือ Blueprint (พิมพ์เขียว) สำหรับสร้าง Object (วัตถุ) ที่มีคุณสมบัติ (Properties) และพฤติกรรม (Methods) เป็นของตัวเอง คล้ายกับแม่พิมพ์ที่ใช้สร้างวัตถุต่าง ๆ

2. Class มีหน้าที่อะไร


*คำตอบ*

หน้าที่ของ Class
-เป็นพิมพ์เขียวสำหรับสร้างอ็อบเจ็กต์
-สนับสนุน OOP เช่น Encapsulation, Inheritance, Polymorphism
-จัดระเบียบโค้ดให้เป็นสัดส่วน
-ทำให้โค้ดสามารถใช้ซ้ำได้ (Reusable)
-ช่วยสร้างระบบที่ขยายได้ง่ายและดูแลง่าย

3. ทำไมเราต้องใช้ Class


*คำตอบ*

Class มีประโยชน์หลายอย่าง: 
-ช่วยจัดระเบียบโค้ด
-สนับสนุนแนวคิด OOP เช่น Encapsulation, Inheritance, Polymorphism
-ทำให้โค้ดนำกลับมาใช้ซ้ำได้ ลดความซ้ำซ้อน
-ช่วยให้ระบบขยายได้ง่าย (Scalability)
-ใช้ร่วมกับ Framework และ Library ได้สะดวก
